```python
#replace each range in age coulmn to its coressponding category ex:0-17 ->child ,18-25 ->teenage, 26-50 -> adult , 51-55+ ->old
```

```python
#1: get all ranges in the data set
df['age'].unique()
```

```python
# 2: map these ranges
map_object = dict()
map_object['0-17'] = 'child'
map_object['18-25'] = 'teenage'
map_object['26-35'] = 'adult'
map_object['36-45'] = 'adult'
map_object['46-50'] = 'adult'
map_object['51-55'] = 'old'
map_object['55+'] = 'old'
df.loc[:,'age'] = df.loc[:,'age'].replace(map_object)
```
