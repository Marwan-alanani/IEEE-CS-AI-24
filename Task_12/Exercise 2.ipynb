### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users_occu_group  = users.groupby('occupation')
pd.DataFrame(users_occu_group['age'].mean()).head()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
percent_occu_gender = users_occu_group['gender'].value_counts(normalize=True)
percent_occu_gender.head()
```

```python
percent_occu_gender = pd.DataFrame(percent_occu_gender)
percent_occu_gender = percent_occu_gender.reset_index(level = ['occupation'])
percent_occu_gender.head()
```

```python
percent_occu_gender.loc['M'].sort_values(by= 'proportion',ascending = False).head()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users_occu_group['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users_occu_gender_group = users.groupby(['occupation','gender'])
pd.DataFrame(users_occu_gender_group['age'].mean())
```

### Step 8.  For each occupation present the percentage of women and men

```python
percent_occu_gender = users_occu_group['gender'].value_counts(normalize=True) * 100
percent_occu_gender
```

```python
perc_per_occu = pd.DataFrame(percent_occu_gender)
perc_per_occu.rename(columns = {'proportion':'percentage'})
```
